```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data = pd.read_csv("C:/Bank loan/merged_data.csv")  
data.head()
```

```python
data.Gender=data.Gender.map({'Male':1,'Female':0})
```

```python
categorical_cols = ['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area']
for col in categorical_cols:
    sns.countplot(x=col, data=data)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
numerical_cols = ['ApplicantIncome', 'CoapplicantIncome', 'LoanAmount', 'Loan_Amount_Term']
for col in numerical_cols:
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.boxplot(y=data['LoanAmount'])
plt.title('Outlier Detection - Loan Amount')
plt.show()
```

```python
property_counts = data['Property_Area'].value_counts()
plt.pie(property_counts, labels=property_counts.index, autopct='%1.1f%%')
plt.title('Property Area Distribution')
plt.axis('equal')
plt.show()
```
